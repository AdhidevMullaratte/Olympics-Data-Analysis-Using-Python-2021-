```python
gold_medals=athletes_df[(athletes_df['Medal']=='Gold')]
gold_medals.head()
```

```python
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
#gold_medals= np.isfinite(gold_medals['Age'])
gold_medals.head()
```

```python
gold_medals['ID'][(gold_medals['Age']>60)].count()
```

```python
Sport_event=gold_medals['Sport'][(gold_medals['Age']>60)]
sports=Sport_event.value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=Sport_event, order=sports.index, palette='Set3')
plt.title('Count of Sports Events for Participants Over 60')
plt.xlabel('Count')
plt.ylabel('Sport')
plt.tight_layout()
plt.show()
```

```python
total_goldmedals=gold_medals.Region.value_counts().reset_index(name=" gold_Medals")
total_goldmedals.head()
```

```python
total_goldmedal= pd.DataFrame({
    'Region': ['USA', 'Russia', 'Germany', 'UK', 'Italy'],
    'gold_Medals': [2627, 1599, 1293, 657, 567]
})

# Create the catplot (bar plot)
plt.figure(figsize=(10, 6))
sns.catplot(data=total_goldmedal, x='Region', y='gold_Medals', kind='bar', palette='Set3')
plt.title('Count of Gold Medals by Region')
plt.xlabel('Region')
plt.ylabel('Gold Medals')
plt.tight_layout()
plt.show()
```

```python
max_year=athletes_df.Year.max()
max_year
```

```python
team=athletes_df[(athletes_df.Year==max_year) & (athletes_df.Medal=='Gold')].Team
team.value_counts().head()
```

```python
team_counts=team.value_counts()
plt.figure(figsize=(20, 15))
team_counts.plot(kind='bar', color='gold')
plt.title('Top 5 Teams with the Most Gold Medals in {}'.format(max_year))
plt.xlabel('Team')
plt.ylabel('Number of Gold Medals')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
