```python
#data loading

region= pd.read_csv("/Users/adhidevmullaratte/Downloads/archive/noc_regions.csv")

athletes= pd.read_csv("/Users/adhidevmullaratte/Downloads/archive/athlete_events.csv")
```

```python
# top 10 countries
top_10_countries =athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# plot for the top ten countries
plt.figure(figsize=(12,6))
plt.title("overall participation from each countries")
sns.barplot(x=top_10_countries.index,y=top_10_countries);
```

```python
# histogram of age of participants
plt.figure(figsize=(12,6))
plt.title("Age distribution of participants")
plt.xlabel("Age")
plt.ylabel("number of participants")
plt.hist(athletes_df.Age,bins= np.arange(10,80,2),color= "blue",edgecolor="black");         
```

```python
wintersports=athletes_df[athletes_df.Season=="Winter"].Sport.unique()
wintersports
```

```python
#gender count
gender_count=athletes_df.Sex.value_counts()
gender_count
```

```python
#piechart
plt.figure(figsize=(12,6))
plt.title("piechart of gender count")
plt.pie(gender_count,labels=gender_count.index,autopct='%1.1f%%',startangle=150);
```
